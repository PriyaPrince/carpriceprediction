### Random Forest Regressor

```python
# Fitting data into RF Model

from sklearn.ensemble import RandomForestRegressor
RF_model = RandomForestRegressor()
```

```python
# Hyperparameter Tuning
```

```python
#Randomized Search CV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python

from sklearn.model_selection import RandomizedSearchCV
```

```python
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
```

```python
rf_random = RandomizedSearchCV(estimator = RF_model, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
rf_random.best_estimator_
```

```python
rf_random.best_params_
```

```python
rf_random.best_score_
```

```python
RF_model = RandomForestRegressor(n_estimators=1000,min_samples_split= 2,min_samples_leaf=1,max_features='sqrt',max_depth=25)
```

```python
RF_model.fit(X_train,y_train)
```

```python
y_pred = RF_model.predict(X_test)
```

```python
from sklearn.metrics import *
```

```python
print('MAE:', mean_absolute_error(y_test, y_pred))
print('MSE:', mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(mean_squared_error(y_test, y_pred)))
```
